## Filtering

### Filtering TGDD

```python
have_ssd=[]
for i in df_tgdd.disk:
    ssd=0
    if (i.find('hỗ trợ khe cắm ssd')!=-1):
        ssd=0
    elif(i.find('ssd')!=-1):
        ssd=1
    else:
        ssd=0
    have_ssd.append(ssd)
new_tgdd = df_tgdd.drop(['link', 'screen_port','id', 'review_count', 'review_point', 'brand'],axis=1)
new_tgdd['category'] = new_tgdd['tag']
new_tgdd['ssd'] = have_ssd
new_tgdd = new_tgdd.drop('tag', axis=1)
```

```python
new_tgdd
```

```python
data_2={
    'name':new_tgdd.title, 'cpu_name':new_tgdd.cpu_tech,'cpu_type':new_tgdd.cpu_type,
   'ram':new_tgdd.ram,'graphic':new_tgdd.graphic_card,'weight':new_tgdd.weight,'ssd':new_tgdd.ssd,
    'battery': new_tgdd.battery,
   'category': new_tgdd.category,
}
df_2=pd.DataFrame(data_2)
df_2
```

### Filtering NguyenKim

```python
for i in range(len(df_nk.Type_Storage)):
    if (type(df_nk.Type_Storage[i])==float):
        df_nk.Type_Storage[i]=str(df_nk.Type_Storage[i])
        
have_ssd_nk=[]
for i in df_nk.Type_Storage:
    ssd=0
    if(i.find('SSD')!=-1):
        ssd=1
    have_ssd_nk.append(ssd)
new_df_nk=df_nk['SSD']=have_ssd_nk
new_df_nk=df_nk.drop(['Type_RAM'],axis=1)
```

```python
new_df_nk_formatted = {
    'name': new_df_nk.Name,
    'cpu_name': new_df_nk.CPU,
    'cpu_type': new_df_nk.Type_CPU,
    'graphic': new_df_nk.Graphic,
    'weight': new_df_nk.Weight,
    'ssd': new_df_nk.SSD,
    'category': new_df_nk.Type_laptop,
    'disk': new_df_nk.Storage
}
new_df_nk_formatted = pd.DataFrame(new_df_nk_formatted)
new_df_nk_formatted
```
