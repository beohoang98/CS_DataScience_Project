## Filtering

### Filtering TGDD

```python
have_ssd=[]
for i in df_tgdd.disk:
    ssd=0
    if (i.find('hỗ trợ khe cắm ssd')!=-1):
        ssd=0
    elif(i.find('ssd')!=-1):
        ssd=1
    else:
        ssd=0
    have_ssd.append(ssd)
new_tgdd=df_tgdd['SSD']=have_ssd
new_tgdd=df_tgdd.drop(['link','screen_port','keyboard_light','Unnamed: 0','id','screen_size','resolution','disk','battery'],axis=1)
```

```python
new_tgdd
```

```python
data_2={'Name':new_tgdd.title,'Price':new_tgdd.price,'CPU':new_tgdd.cpu_tech,'Type_CPU':new_tgdd.cpu_type,
       'RAM':new_tgdd.ram,'Graphic':new_tgdd.graphic_card,'Weight':new_tgdd.weight,'Score_star':new_tgdd.review_point,
        'Vote':new_tgdd.review_count,'Brand':new_tgdd.brand	,'SSD':new_tgdd.SSD}
df_2=pd.DataFrame(data_2)
df_2
```

### Filtering NguyenKim

```python
for i in range(len(df_nk.Type_Storage)):
    if (type(df_nk.Type_Storage[i])==float):
        df_nk.Type_Storage[i]=str(df_nk.Type_Storage[i])
        
have_ssd_nk=[]
for i in df_nk.Type_Storage:
    ssd=0
    if(i.find('SSD')!=-1):
        ssd=1
    have_ssd_nk.append(ssd)
new_df_nk=df_nk['SSD']=have_ssd_nk
new_df_nk=df_nk.drop(['Storage','Type_Storage','Type_RAM'],axis=1)
```

```python
new_df_nk
```

```python
new_df_nk.to_csv('New_NK.csv',index=0)
```

### Filtering Tiki

```python
data_tiki_filtered = {
    'Name': df_tiki.title,
    'Price': df_tiki.price,
    'CPU': df_tiki.cpu,
    'Type_CPU': df_tiki.type_cpu,
    'RAM': df_tiki.ram,
    'Graphic': df_tiki.graphic,
    'Weight': df_tiki.weight,
    'Score_star': df_tiki.rating_average,
    'Vote': df_tiki.reviews_count,
    'Brand': df_tiki.brand,
    'SSD': df_tiki.ssd.apply(lambda ssd: 1 if ssd else 0)
}
df_tiki_filtered = pd.DataFrame(data_tiki_filtered)
```

## Merging (Tron 3 dataframe vo thanh 1)

```python
df = pd.read_csv('New_NK.csv')
merge = df.append(df_2, ignore_index=True)
merge = df.append(df_tiki_filtered, ignore_index=True)
merge.Weight = merge.Weight.apply(lambda w: re.sub(r'[ a-zA-Z]', '', w) if isinstance(w, str) else w)

merge = merge.loc[:, ~merge.columns.str.contains('^Unnamed')]

merge
```

```python
merge.to_csv('Merge.csv', index=False)
```
